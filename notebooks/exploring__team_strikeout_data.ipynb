```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

# Configurações para melhor visualização
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")
plt.rcParams['figure.figsize'] = (12, 8)
plt.rcParams['font.size'] = 10

# Configurando para exibir todas as colunas
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)

print("✅ Bibliotecas importadas e configurações aplicadas")
```

```python
# Carregando os dados
df = pd.read_csv('../team_strikeout_percentage.csv')

# Exibindo informações básicas
print("=== INFORMAÇÕES BÁSICAS ===")
print(f"Dimensões do dataset: {df.shape}")
print(f"\nPrimeiras 5 linhas:")
print(df.head())

print(f"\nInformações do dataset:")
print(df.info())

print(f"\nEstatísticas descritivas:")
print(df.describe())

# =============================================================================
# CÉLULA 3: VERIFICAÇÃO DE QUALIDADE DOS DADOS
# =============================================================================

# Verificando se há valores nulos
print("=== VERIFICAÇÃO DE DADOS NULOS ===")
print(df.isnull().sum())

# Verificando valores únicos por coluna
print(f"\n=== VALORES ÚNICOS ===")
for col in df.columns:
    print(f"{col}: {df[col].nunique()} valores únicos")

# Verificando se há valores duplicados
print(f"\nLinhas duplicadas: {df.duplicated().sum()}")
```

```python
# =============================================================================
# CÉLULA 4: ANÁLISE ESTATÍSTICA POR ANO
# =============================================================================

# Análise estatística por ano
print("=== ANÁLISE POR ANO ===")
anos = ['2023', '2024', '2025']

for ano in anos:
    print(f"\n{ano}:")
    print(f"  Média: {df[ano].mean():.2f}%")
    print(f"  Mediana: {df[ano].median():.2f}%")
    print(f"  Desvio padrão: {df[ano].std():.2f}%")
    print(f"  Mínimo: {df[ano].min():.2f}%")
    print(f"  Máximo: {df[ano].max():.2f}%")
    print(f"  Range: {df[ano].max() - df[ano].min():.2f}%")

# =============================================================================
# CÉLULA 5: VISUALIZAÇÃO 1 - DISTRIBUIÇÃO POR ANO
# =============================================================================

# Visualização 1: Distribuição por ano
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

for i, ano in enumerate(anos):
    axes[i].hist(df[ano], bins=15, alpha=0.7, edgecolor='black')
    axes[i].axvline(df[ano].mean(), color='red', linestyle='--', label=f'Média: {df[ano].mean():.1f}%')
    axes[i].axvline(df[ano].median(), color='green', linestyle='--', label=f'Mediana: {df[ano].median():.1f}%')
    axes[i].set_title(f'Distribuição de Strikeout % - {ano}')
    axes[i].set_xlabel('Porcentagem de Strikeout')
    axes[i].set_ylabel('Frequência')
    axes[i].legend()
    axes[i].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# =============================================================================
# CÉLULA 6: VISUALIZAÇÃO 2 - BOXPLOT COMPARATIVO
# =============================================================================

# Visualização 2: Boxplot comparativo
plt.figure(figsize=(10, 6))
df_melted = df.melt(id_vars=['Team'], value_vars=anos, var_name='Ano', value_name='Strikeout_Percentage')

sns.boxplot(data=df_melted, x='Ano', y='Strikeout_Percentage')
plt.title('Comparação da Distribuição de Strikeout % por Ano')
plt.ylabel('Porcentagem de Strikeout')
plt.grid(True, alpha=0.3)
plt.show()

# Adicionando pontos individuais
plt.figure(figsize=(10, 6))
sns.boxplot(data=df_melted, x='Ano', y='Strikeout_Percentage')
sns.stripplot(data=df_melted, x='Ano', y='Strikeout_Percentage', color='red', alpha=0.5, size=4)
plt.title('Distribuição de Strikeout % por Ano (com pontos individuais)')
plt.ylabel('Porcentagem de Strikeout')
plt.grid(True, alpha=0.3)
plt.show()

# =============================================================================
# CÉLULA 7: VISUALIZAÇÃO 3 - EVOLUÇÃO TEMPORAL POR EQUIPE
# =============================================================================

# Visualização 3: Evolução temporal por equipe
plt.figure(figsize=(16, 10))

# Ordenando equipes pela média de 2023-2025
df['media_geral'] = df[anos].mean(axis=1)
df_sorted = df.sort_values('media_geral', ascending=False)

for i, team in enumerate(df_sorted['Team']):
    values = df_sorted.loc[df_sorted['Team'] == team, anos].values[0]
    plt.plot(anos, values, marker='o', linewidth=2, markersize=6, label=team)

plt.title('Evolução da Porcentagem de Strikeout por Equipe (2023-2025)', fontsize=14)
plt.xlabel('Ano', fontsize=12)
plt.ylabel('Porcentagem de Strikeout', fontsize=12)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=8)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# =============================================================================
# CÉLULA 8: ANÁLISE DAS EQUIPES COM MAIOR E MENOR PORCENTAGEM
# =============================================================================

# Análise das equipes com maior e menor porcentagem por ano
print("=== EQUIPES COM MAIOR E MENOR STRIKEOUT % ===")

for ano in anos:
    print(f"\n{ano}:")
    
    # Maior porcentagem
    max_team = df.loc[df[ano].idxmax(), 'Team']
    max_value = df[ano].max()
    print(f"  Maior: {max_team} ({max_value:.2f}%)")
    
    # Menor porcentagem
    min_team = df.loc[df[ano].idxmin(), 'Team']
    min_value = df[ano].min()
    print(f"  Menor: {min_team} ({min_value:.2f}%)")
    
    # Diferença
    print(f"  Diferença: {max_value - min_value:.2f}%")

# =============================================================================
# CÉLULA 9: ANÁLISE DE CORRELAÇÃO ENTRE ANOS
# =============================================================================

# Análise de correlação entre anos
print("=== CORRELAÇÃO ENTRE ANOS ===")
correlation_matrix = df[anos].corr()
print(correlation_matrix)

# Visualização da matriz de correlação
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0, 
            square=True, linewidths=0.5, cbar_kws={'shrink': 0.8})
plt.title('Matriz de Correlação entre Anos')
plt.tight_layout()
plt.show()

# =============================================================================
# CÉLULA 10: ANÁLISE DE MUDANÇA ANO A ANO
# =============================================================================

# Análise de mudança ano a ano
print("=== ANÁLISE DE MUDANÇA ANO A ANO ===")

# Calculando mudanças
df['mudanca_2023_2024'] = df['2024'] - df['2023']
df['mudanca_2024_2025'] = df['2025'] - df['2024']
df['mudanca_total'] = df['2025'] - df['2023']

print("\nMudanças 2023-2024:")
print(f"Média: {df['mudanca_2023_2024'].mean():.2f}%")
print(f"Equipe com maior aumento: {df.loc[df['mudanca_2023_2024'].idxmax(), 'Team']} ({df['mudanca_2023_2024'].max():.2f}%)")
print(f"Equipe com maior queda: {df.loc[df['mudanca_2023_2024'].idxmin(), 'Team']} ({df['mudanca_2023_2024'].min():.2f}%)")

print("\nMudanças 2024-2025:")
print(f"Média: {df['mudanca_2024_2025'].mean():.2f}%")
print(f"Equipe com maior aumento: {df.loc[df['mudanca_2024_2025'].idxmax(), 'Team']} ({df['mudanca_2024_2025'].max():.2f}%)")
print(f"Equipe com maior queda: {df.loc[df['mudanca_2024_2025'].idxmin(), 'Team']} ({df['mudanca_2024_2025'].min():.2f}%)")

print("\nMudanças totais 2023-2025:")
print(f"Média: {df['mudanca_total'].mean():.2f}%")
print(f"Equipe com maior aumento: {df.loc[df['mudanca_total'].idxmax(), 'Team']} ({df['mudanca_total'].max():.2f}%)")
print(f"Equipe com maior queda: {df.loc[df['mudanca_total'].idxmin(), 'Team']} ({df['mudanca_total'].min():.2f}%)")

# =============================================================================
# CÉLULA 11: VISUALIZAÇÃO DAS MUDANÇAS
# =============================================================================

# Visualização das mudanças
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Mudanças 2023-2024
axes[0].bar(range(len(df)), df['mudanca_2023_2024'].sort_values(ascending=False))
axes[0].set_title('Mudanças 2023-2024')
axes[0].set_ylabel('Mudança na % de Strikeout')
axes[0].axhline(y=0, color='black', linestyle='-', alpha=0.3)
axes[0].grid(True, alpha=0.3)

# Mudanças 2024-2025
axes[1].bar(range(len(df)), df['mudanca_2024_2025'].sort_values(ascending=False))
axes[1].set_title('Mudanças 2024-2025')
axes[1].set_ylabel('Mudança na % de Strikeout')
axes[1].axhline(y=0, color='black', linestyle='-', alpha=0.3)
axes[1].grid(True, alpha=0.3)

# Mudanças totais
axes[2].bar(range(len(df)), df['mudanca_total'].sort_values(ascending=False))
axes[2].set_title('Mudanças Totais 2023-2025')
axes[2].set_ylabel('Mudança na % de Strikeout')
axes[2].axhline(y=0, color='black', linestyle='-', alpha=0.3)
axes[2].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# =============================================================================
# CÉLULA 12: RANKING DAS EQUIPES POR MÉDIA GERAL
# =============================================================================

# Ranking das equipes por média geral
print("=== RANKING GERAL POR MÉDIA (2023-2025) ===")
ranking_geral = df[['Team', 'media_geral']].sort_values('media_geral', ascending=False)
ranking_geral['Ranking'] = range(1, len(ranking_geral) + 1)
ranking_geral = ranking_geral[['Ranking', 'Team', 'media_geral']]
ranking_geral['media_geral'] = ranking_geral['media_geral'].round(2)

print(ranking_geral.to_string(index=False))

# Top 10 e Bottom 10
print("\n=== TOP 10 EQUIPES ===")
print(ranking_geral.head(10).to_string(index=False))

print("\n=== BOTTOM 10 EQUIPES ===")
print(ranking_geral.tail(10).to_string(index=False))

# =============================================================================
# CÉLULA 13: ANÁLISE DE CONSISTÊNCIA
# =============================================================================

# Análise de consistência (equipes com menor variação)
print("=== ANÁLISE DE CONSISTÊNCIA ===")
df['desvio_padrao'] = df[anos].std(axis=1)
df['variacao'] = df[anos].max(axis=1) - df[anos].min(axis=1)

print("\nEquipes mais consistentes (menor desvio padrão):")
consistencia = df[['Team', 'desvio_padrao', 'variacao']].sort_values('desvio_padrao')
print(consistencia.head(10).to_string(index=False))

print("\nEquipes menos consistentes (maior desvio padrão):")
print(consistencia.tail(10).to_string(index=False))

# Visualização da consistência
plt.figure(figsize=(12, 8))
plt.scatter(df['media_geral'], df['desvio_padrao'], alpha=0.7, s=100)

# Adicionando labels das equipes
for i, team in enumerate(df['Team']):
    plt.annotate(team, (df['media_geral'].iloc[i], df['desvio_padrao'].iloc[i]), 
                xytext=(5, 5), textcoords='offset points', fontsize=8)

plt.xlabel('Média Geral de Strikeout %')
plt.ylabel('Desvio Padrão')
plt.title('Consistência vs Performance Média')
plt.grid(True, alpha=0.3)
plt.show()

# =============================================================================
# CÉLULA 14: RESUMO FINAL
# =============================================================================

# Resumo final
print("=== RESUMO FINAL ===")
print(f"\nDataset analisado: {df.shape[0]} equipes, {df.shape[1]-1} anos")
print(f"Período: 2023-2025")

print(f"\nTendências gerais:")
for i, ano in enumerate(anos):
    print(f"  {ano}: Média de {df[ano].mean():.2f}% de strikeout")

print(f"\nEquipe com maior média geral: {ranking_geral.iloc[0]['Team']} ({ranking_geral.iloc[0]['media_geral']}%)")
print(f"Equipe com menor média geral: {ranking_geral.iloc[-1]['Team']} ({ranking_geral.iloc[-1]['media_geral']}%)")

print(f"\nEquipe mais consistente: {consistencia.iloc[0]['Team']} (desvio: {consistencia.iloc[0]['desvio_padrao']:.2f}%)")
print(f"Equipe menos consistente: {consistencia.iloc[-1]['Team']} (desvio: {consistencia.iloc[-1]['desvio_padrao']:.2f}%)")

print(f"\nCorrelação entre 2023-2024: {correlation_matrix.loc['2023', '2024']:.3f}")
print(f"Correlação entre 2024-2025: {correlation_matrix.loc['2024', '2025']:.3f}")
print(f"Correlação entre 2023-2025: {correlation_matrix.loc['2023', '2025']:.3f}")

print("\n✅ Análise concluída com sucesso!") 
```
